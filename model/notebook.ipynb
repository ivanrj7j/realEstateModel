# House Price Recommendation 

This is a model that takes in data from makaan.com and recommends the price of the house using Linear Regression.
This can be used when the user is entering the detials of the house and the system can automatically predict the price

## Analysing data

```python
import pandas as pd
from matplotlib import pyplot as plt
# importing matplotlib and pandas 
```

```python
housingData = pd.read_csv('data.csv')
# importing the data 
```

```python
housingData.hist(figsize=(12,15))
# histogram of all data points 
```

> ### Observations
> - The Age of all houses are around 0 to 25 years, 1 being the age of most
> - Amneties score of all houses are around 1 to 10 0 being prominent (showing the lack of luxurious features like swimming pools) 
> - The area of most house are around 2000 square feats(this observation might change when the description of the data is observed)

## Making necessary changes to data

```python
Features = housingData
Labels = housingData['price']
Features.pop('price')
# getting the features and labels 
```

```python
Features.head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(Features, Labels)
```

```python
import pickle

with open('model.sav', "wb") as file:
    pickle.dump(model, file)
```
